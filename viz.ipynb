```python
import matplotlib.pyplot as plt 
from generator import build_adj
import numpy as np 

for i in range(2,10):
    ax = plt.gca()
    mat = build_adj(i)
    ax.set_title(f'2^{i} Vertices')
    ax.imshow(mat, vmin=0, vmax=1, extent=[0,2**i, 0,2**i])
    ax.set_xticks(np.arange(0,2**i, 2**(i-2)))
    ax.set_yticks(np.arange(0,2**i, 2**(i-2)))
    ax.grid()

    print(sum([mat[i][i] for i in range(mat.size(0))]))
    plt.savefig(f'img/2_pow_{i}')
    plt.show()
```
